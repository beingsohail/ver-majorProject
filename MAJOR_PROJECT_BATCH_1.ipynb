### Feature Selection

```python
df.columns
```

```python
# Here the columns used in 'X' are the independent variables.
# and the column named 'Survived' is the dependent variable.

X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']].values

Y = df[['Survived']].values
```

### Modelling

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Support vector machine

```python
from sklearn.svm import SVC
```

```python
cls = SVC()
```

```python
cls.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = cls.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dsc = DecisionTreeClassifier()
```

```python
dsc.fit(x_train, y_train)
```

```python
y_pred = dsc.predict(x_test)
```

```python
accuracy_score(y_pred, y_test)
```

### Question 1)  Which gender is more likely to survived? 

```python
df['gender'] = df.Sex.map({0:'male', 1:'female'})

survived = df[df['Survived']==1]['gender'].value_counts()
dead = df[df['Survived']==0]['gender'].value_counts()
df1 = pd.DataFrame([survived,dead])
df1.index = ['Survived','Dead']
df1.plot(kind='bar',stacked=True, figsize=(10,5))
```

###### The Chart confirms Women more likely survivied than Men

### Question 2) Which class was more likely to survive than other classes?

```python
survived = df[df['Survived']==1]['Pclass'].value_counts()
dead = df[df['Survived']==0]['Pclass'].value_counts()
df1 = pd.DataFrame([survived,dead])
df1.index = ['Survived','Dead']
df1.plot(kind='bar',stacked=True, figsize=(10,5))
```
